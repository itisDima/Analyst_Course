### Задача №4.
Как вы уже могли заметить, при сборе датасета произошла небольшая ошибка, и названия колонок были записаны в неудобном для дальнейшей работы формате. Исправьте ситуацию, приведя названия столбцов к нижнему регистру и заменив пробелы на знак нижнего подчеркивания. 
**Например: Is Canceled -> is_canceled**

```python
# Функция, котороя приводит названия столбцов к нижнему регистру и меняет пробел между словами на _.
def adsc(name):
    new_name=name.lower().replace(' ', '_') 
    return new_name                         
```

```python
bookings = bookings.rename(columns=adsc) 
bookings.head(5)
```

### Задача №5.
Пользователи из каких стран совершили наибольшее число успешных бронирований? Бронирование считается успешным, если в дальнейшем не было отменено (переменная is_canceled). В качестве ответа выберите страны, входящие в **топ-5**.

```python
bookings.query("reservation_status == 'Check-Out'").groupby('country', as_index = False) \
        .agg({'reservation_status':'count'}) \
        .sort_values('reservation_status', ascending=False).head(5)
```

## Задача №6.
 На сколько ночей (stays_total_nights)  в среднем бронируют отели типа **City Hotel**? **Resort Hotel**? Запишите полученные значения в пропуски с точностью до 2 знаков после точки.

```python
bookings.groupby('hotel', as_index = False) \
          .agg({'stays_total_nights':'mean'}).rename(columns={'stays_total_nights':'mean_total_nights'}) \
          .sort_values('mean_total_nights', ascending=False).round(2) 
```

### Задача №8.
Теперь проанализируйте даты запланированного прибытия (arrival_date_year).

На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?

1. В 2017 чаще всего бронировали отель на октябрь
2. В 2016 чаще всего бронировали отель на май
3. В 2016 чаще всего оформляли бронирование на сентябрь
4. В 2016 чаще всего бронировали отель на октябрь
5. В 2017 чаще всего оформляли бронирование на май

```python
bookings.groupby(['arrival_date_month', 'arrival_date_year'], as_index=False) \
        .agg({'reservation_status_date' : 'count'}) \
        .sort_values('arrival_date_month', ascending=False) \
        .rename(columns = {'reservation_status_date':'count_reservation_by_date'})
```

Ответ:

4. В 2016 чаще всего бронировали отель на октябрь

5. В 2017 чаще всего оформляли бронирование на май


### Задача №9.
Сгруппируйте данные по годам, а затем проверьте, на какой месяц (arrival_date_month) бронирования отеля типа **City Hotel** отменялись чаще всего в 2015? 2016? 2017? 

```python
bookings.query("is_canceled == 1 and hotel == 'City Hotel'") \
        .groupby(['arrival_date_month', 'arrival_date_year'], as_index=False) \
        .agg({'reservation_status' : 'count'}) \
        .sort_values(['arrival_date_year', 'reservation_status'], ascending=False)
```

Ответ:

_Май_ - 2017

_Октябрь_ - 2016

_Сентябрь_ - 2015

### Задача №11.
Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

* City hotel – отель находится в городе
* Resort hotel – отель курортный

В качестве ответа укажите наибольшее среднее total_kids, округлив до 2 знаков после точки

```python
bookings['total_kids'] = bookings.children + bookings.babies
```

```python
bookings.groupby('hotel', as_index=False) \
        .agg({'total_kids' : 'mean'}) \
        .sort_values('total_kids', ascending=False) \
        .rename(columns = {'total_kids':'mean_total_kids'}).round(2)
```

**Ответ :** наибольшее среднее значение переменной total_kids равно 0.14 для Resort Hotel

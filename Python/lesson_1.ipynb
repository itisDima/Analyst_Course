```python
def adsc(name):
    new_name=name.lower().replace(' ', '_') #Создал функцию, которая меняет большие буквы на маленькие 
    return new_name                         # и делает замену пробелов на _
```

```python
bookings = bookings.rename(columns=adsc) 
bookings
```

```python
bookings.query("reservation_status == 'Check-Out'").groupby('country') \
        .agg({'reservation_status':'count'}) \
        .sort_values('reservation_status', ascending=False)
```

```python
p=bookings.groupby('hotel') \
          .agg({'stays_total_nights':'mean'}) \
          .sort_values('stays_total_nights', ascending=False) 
p.round(2) 
```

```python
bookings.query('assigned_room_type != reserved_room_type') 
```

```python
#На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?
bookings.groupby(['arrival_date_month', 'arrival_date_year']) \
        .agg({'reservation_status_date' : 'count'}) \
        .sort_values('arrival_date_month', ascending=False)
```

```python
#Сгруппируйте данные по годам, а затем проверьте, на какой месяц 
#бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 

bookings.query("is_canceled == 1 and hotel == 'City Hotel'") \
        .groupby(['arrival_date_month', 'arrival_date_year'], as_index=False) \
        .agg({'reservation_status' : 'count'}) \
        .sort_values('reservation_status', ascending=False)
```

```python
#Посмотрите на числовые характеристики трёх колонок: adults, children и babies.
#Какая из них имеет наибольшее среднее значение?

bookings[{'adults', 'children', 'babies'}].describe()
```

**Ответ :** abults с средним значением 1.856403

```python
#Создайте колонку total_kids, объединив столбцы children и babies. 
#Для отелей какого типа среднее значение переменной оказалось наибольшим?
#В качестве ответа укажите наибольшее среднее total_kids, округлив до 2 знаков после точки.

bookings['total_kids'] = bookings.children + bookings.babies
```

```python
bookings.groupby('hotel', as_index=False) \
        .agg({'total_kids' : 'mean'}) \
        .sort_values('total_kids', ascending=False) \
        .round(2)
```

**Ответ :** наибольшее среднее значение переменной total_kids равно 0.14 для Resort Hotel
